# Feature Detection and Descriptors

ORB, SIFT (if available), feature matching, RANSAC for homography.

### Example: ORB feature detection


```
import cv2
img = cv2.imread('datasets/sample_images/box1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
orb = cv2.ORB_create()
kp = orb.detect(gray, None)
img2 = cv2.drawKeypoints(img, kp, None, color=(0,255,0))
```
