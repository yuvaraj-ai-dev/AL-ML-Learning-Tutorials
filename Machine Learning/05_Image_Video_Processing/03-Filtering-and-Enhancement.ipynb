# Filtering and Enhancement

Linear filters, Gaussian, median, bilateral, non-local means.

### Example: remove salt-and-pepper noise with median filter


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('datasets/sample_images/noisy_sp.png',0)
d = cv2.medianBlur(img,5)
plt.subplot(1,2,1); plt.imshow(img,cmap='gray'); plt.title('Noisy'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(d,cmap='gray'); plt.title('Median'); plt.axis('off')
```
