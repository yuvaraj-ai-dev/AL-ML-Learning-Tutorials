# Video Processing Basics

Reading/writing video, frame iteration, timestamps.

### Example: extract frames


```
import cv2
cap = cv2.VideoCapture('datasets/sample_videos/sample_clip.mp4')
count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    if count % 10 == 0:
        cv2.imwrite(f'datasets/frames/frame_{count:04d}.png', frame)
    count += 1
cap.release()
```
