## What is Sentiment Analysis?Determines if text expresses:- **Positive**: Happy, satisfied, good- **Negative**: Angry, disappointed, bad- **Neutral**: Factual, no emotion**Applications**:- Product reviews- Customer feedback- Social media monitoring- Stock market analysis

```python
from textblob import TextBlobprint('='*60)print('SENTIMENT ANALYSIS EXAMPLE')print('='*60)texts = [    'I absolutely love this product!',    'This is terrible and useless.',    'The weather is sunny.']for text in texts:    blob = TextBlob(text)    print(f'\nText: "{text}"')    print(f'  Polarity: {blob.sentiment.polarity:.2f} (-1=negative, 0=neutral, 1=positive)')    print(f'  Subjectivity: {blob.sentiment.subjectivity:.2f} (0=objective, 1=subjective)')        if blob.sentiment.polarity > 0.5:        sentiment = 'POSITIVE'    elif blob.sentiment.polarity < -0.5:        sentiment = 'NEGATIVE'    else:        sentiment = 'NEUTRAL'    print(f'  Sentiment: {sentiment}')
```
