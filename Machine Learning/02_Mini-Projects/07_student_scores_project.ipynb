# Mini-project: Student Score Analyzer (Solution)

Compute averages, find top student, and plot subject averages.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'student':['John','Sara','Lee','Maya','Ezra'],
    'math':[78,92,85,88,74],
    'science':[85,88,79,90,82],
    'english':[80,91,84,86,78]
})
print(data)
```

## Compute average per student and per subject

```python
data['avg'] = data[['math','science','english']].mean(axis=1)
print(data[['student','avg']])

subject_avg = data[['math','science','english']].mean()
print('\nSubject averages:\n', subject_avg)
```

## Top scoring student

```python
top = data.sort_values('avg', ascending=False).iloc[0]
print('Top student:', top['student'], 'avg', top['avg'])
```

## Plot subject averages

```python
subject_avg.plot.bar(title='Subject averages')
plt.ylabel('Average score')
plt.tight_layout()
plt.show()
```
