# Mini-project: Weather Charts (Solution)

Create line, scatter and histogram from fake weather data.

```python
import pandas as pd
import matplotlib.pyplot as plt

days = list(range(1,11))
temp = [18,20,19,21,22,20,18,17,19,21]
humidity = [40,42,50,45,43,41,39,38,44,46]
df = pd.DataFrame({'day':days,'temp':temp,'humidity':humidity})
print(df)
```

## Line plot: temperature over days

```python
plt.figure()
plt.plot(df['day'], df['temp'], marker='o')
plt.title('Temperature over 10 days')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```

## Scatter: humidity vs temperature

```python
plt.figure()
plt.scatter(df['temp'], df['humidity'])
plt.title('Humidity vs Temperature')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.show()
```

## Histogram: temperature distribution

```python
plt.figure()
plt.hist(df['temp'], bins=5)
plt.title('Temperature distribution')
plt.show()
```

## Save a chart

```python
out = './images/weather_temp_line.png'
plt.figure()
plt.plot(df['day'], df['temp'])
plt.savefig(out)
plt.close()
print('Saved', out)
```
