# Mini-project: Simple Image Manipulation (Solution)

**Tools:** NumPy, Matplotlib

**What you'll learn:** array slicing, image representation, plotting images.

## Step 1 — Create a 100x100 grayscale image (array)
We represent images as 2D arrays with values 0-255.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create random grayscale image
img = np.random.randint(0, 256, size=(100,100), dtype=np.uint8)
print('shape:', img.shape, 'dtype:', img.dtype)
```

## Step 2 — Modify top half to white (255)

```python
# Make a copy and modify top half
img2 = img.copy()
img2[:50, :] = 255

# Show original and modified side-by-side
fig, axes = plt.subplots(1,2, figsize=(6,3))
axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('Original')
axes[0].axis('off')
axes[1].imshow(img2, cmap='gray', vmin=0, vmax=255)
axes[1].set_title('Top half white')
axes[1].axis('off')
plt.tight_layout()
plt.show()
```

## Explanation
- Images are arrays. Slicing `img2[:50, :]` selects the top 50 rows (half of 100).
- Setting them to 255 makes them white (grayscale).

## Save result

```python
import os
out = './image_top_white.png'
plt.imsave(out, img2, cmap='gray')
print('Saved to', out)
assert os.path.exists(out)
```
