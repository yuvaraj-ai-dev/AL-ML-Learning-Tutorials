# Mini-project: Sales Dataset Analyzer (Solution)

Create totals, group by category, and plot revenue.

```
import pandas as pd
import matplotlib.pyplot as plt

# Sample sales data
data = pd.DataFrame({
    'product': ['A','B','C','D','E'],
    'price': [10,15,8,20,5],
    'quantity': [3,2,5,1,10],
    'category': ['food','tools','food','tools','food']
})
print(data)
```

## Compute revenue and top product

```
data['revenue'] = data['price'] * data['quantity']
print(data[['product','revenue']])

# Top-selling by revenue
top = data.sort_values('revenue', ascending=False).iloc[0]
print('Top product:', top['product'], 'revenue', top['revenue'])
```

## Group revenue by category and plot

```
grouped = data.groupby('category')['revenue'].sum().reset_index()
print(grouped)

# Bar plot
ax = grouped.plot.bar(x='category', y='revenue', legend=False, title='Revenue by category')
ax.set_ylabel('Revenue')
plt.tight_layout()
plt.show()
```

## Save results

```
out = './sales_revenue_by_category.png'
plt.savefig(out, dpi=120)
print('Saved to', out)
```
