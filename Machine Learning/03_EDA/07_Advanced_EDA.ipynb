## 1. Feature Engineering
Create new meaningful features from existing ones.

```python
import pandas as pd, seaborn as sns

df = sns.load_dataset('titanic').copy()
# Example engineered feature

df['family_size'] = df['sibsp'] + df['parch']
df[['sibsp','parch','family_size']].head()
```

## 2. PCA (Dimensionality Reduction)

PCA (Principal Component Analysis) is a dimensionality reduction technique used in data analysis and machine learning. It transforms a large set of variables into a smaller one that still contains most of the information in the large set.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns, pandas as pd

df = sns.load_dataset('titanic').copy().dropna(subset=['age','fare','sibsp','parch'])
X = df[['age','fare','sibsp','parch']]
X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca[:5]
```

## 3. Multivariate Analysis
Use pairplots, correlation matrices, jointplots.
